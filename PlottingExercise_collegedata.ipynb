# Visualizing Earnings Based On College Majors

In this guided project, we'll explore how using the pandas plotting functionality along with the Jupyter notebook interface allows us to explore data quickly using visualizations.

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.

Using visualizations, we can start to explore questions from the dataset like:

- Do students in more popular majors make more money?
    - Using scatter plots
- How many majors are predominantly male? Predominantly female?
    - Using histograms
- Which category of majors have the most students?
    - Using bar plots

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print("Rows including NA:{}\nRows excluding NA:{}"
      .format(raw_data_count, cleaned_data_count))
```

```python
fig, axes = plt.subplots(3,2, figsize=(10,16))

recent_grads.plot(x='Sample_size', y='Median', kind='scatter'
                  , title='Sample_size vs. Median', ax=axes[0,0])

recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter'
                  , title='Sample_size vs. Unemployment_rate', ax=axes[0,1] )

recent_grads.plot(x='Full_time', y='Median', kind='scatter'
                  , title='Full_time vs. Median', ax=axes[1,0] )

recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter'
                  , title='ShareWomen vs. Unemployment_rate', ax=axes[1,1] )

recent_grads.plot(x='Men', y='Median', kind='scatter'
                  , title='Men vs. Median', ax=axes[2,0] )

recent_grads.plot(x='Women', y='Median', kind='scatter'
                  , title='Women vs. Median', ax=axes[2,1] )
```

```python
fig, axes = plt.subplots(4,2, figsize=(10,16))

recent_grads['Sample_size'].hist(bins=25, range=(0,5000), ax=axes[0,0])
axes[0,0].set_title('Sample_size')
recent_grads['Median'].hist(bins=25, range=(0,120000), ax=axes[0,1])
axes[0,1].set_title('Median')
recent_grads['Employed'].hist(bins=25, range=(0,100000), ax=axes[1,0])
axes[1,0].set_title('Employed')
recent_grads['Full_time'].hist(bins=25, range=(0,10000), ax=axes[1,1])
axes[1,1].set_title('Full_time')
recent_grads['ShareWomen'].hist(bins=25, range=(0,1), ax=axes[2,0])
axes[2,0].set_title('ShareWomen')
recent_grads['Unemployment_rate'].hist(bins=25, range=(0,1), ax=axes[2,1])
axes[2,1].set_title('Unemployment_rate')
recent_grads['Men'].hist(bins=25, range=(0,50000), ax=axes[3,0])
axes[3,0].set_title('Men')
recent_grads['Women'].hist(bins=25, range=(0,50000), ax=axes[3,1])
axes[3,1].set_title('Women')
```
